```python
#Number of crimes committed in each district per hour
crime['count']=1
Dist_hour=pd.pivot_table(crime,
                         values='count',
                         index=['District','Hour of the Day'],
                         aggfunc=sum)
Dist_hour=Dist_hour.reset_index()

dist_sum = pd.DataFrame(Dist_hour.groupby(['District'])['count'].agg('sum'))

dist_sum = dist_sum.reset_index()
Dist_hour = Dist_hour.merge(dist_sum, left_on = 'District', right_on = 'District', how = 'inner')

Dist_hour['Percentage']=(Dist_hour['count_x']/Dist_hour['count_y'])*100
Dist_hour
```
