```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
adidas = pd.read_csv('reviewsAdidas.csv')
asics = pd.read_csv('reviewsAsics.csv')
newbalance = pd.read_csv('reviewsNewBalance.csv')
nike = pd.read_csv('reviewsNike.csv', encoding='ISO-8859-1')
alldata = pd.read_csv('mergeddata.csv', encoding='ISO-8859-1')
```

## adidas

```python
adidas['tokens'] = adidas['review'].apply(str.lower)
adidas['tokens'] = adidas['tokens'].map(nltk.word_tokenize)
adidas.head()
```

```python
adidaslist = [item for sublist in adidas['tokens'].tolist() for item in sublist]
```

```python
import numpy as np

stopset = set(stopwords.words('english'))
stopset.update(['shoe', '.', ',', '"', "'", '?', '!', ':', ';', '<', '>', '(', ')', '[', ']', '{', '}'])
adidastokens = [w for w in adidaslist if not w in stopset]

adidaslemma = []
lemma = WordNetLemmatizer()

for i in adidastokens:
    new = lemma.lemmatize(i)
    adidaslemma.append(new)
    
adidastokensdf = pd.DataFrame(np.array(adidaslemma))
adidastokensdf.columns = ['words']
```

```python
adidaswordlist = adidastokensdf['words'].value_counts()
```

```python
adidasreplace = pd.read_csv("adidaswordlist.csv")
adidasreplace.head()
```

```python
replace = adidasreplace['ATTRIBUTE'].tolist()
search = adidasreplace['WORD'].tolist()
replaceDict = dict(zip(search, replace))

def replaceWord(l):
    new_list = []
    for x in l:
        if x not in list(replaceDict.keys()):
            x = x
        else:
            x = replaceDict[x]
        new_list.append(x)
    return new_list
    
adidas['finaltokens'] = adidas['tokens'].apply(replaceWord)
adidas.head()
```

```python
flatattributes = [item for sublist in adidas['finaltokens'].tolist() for item in sublist]

stopset = set(stopwords.words('english'))
stopset.update(['shoe', 'shoes','.', ',', '"', "'", '?', '!', ':', ';', '<', '>', '(', ')', '[', ']', '{', '}'])
adidastokens = [w for w in flatattributes if not w in stopset]

adidaslemma = []
lemma = WordNetLemmatizer()

for i in adidastokens:
    new = lemma.lemmatize(i)
    adidaslemma.append(new)
    
adidastokensdf = pd.DataFrame(np.array(adidaslemma))
adidastokensdf.columns = ['words']
adidaswordlist = adidastokensdf['words'].value_counts()
adidaswordlist[:20]

adidaswordlist.to_csv("finalorderadidas.csv")
```

## Calculating lifts

```python
alldata['tokens'] = alldata['review'].apply(unicode.lower)
alldata['tokens'] = alldata['tokens'].map(nltk.word_tokenize)
alldata.head()

alldatalist = [item for sublist in alldata['tokens'].tolist() for item in sublist]
```

```python
import numpy as np

stopset = set(stopwords.words('english'))
stopset.update(['shoe', '.', ',', '"', "'", '?', '!', ':', ';', '<', '>', '(', ')', '[', ']', '{', '}'])
alldatatokens = [w for w in alldatalist if not w in stopset]

alldatalemma = []
lemma = WordNetLemmatizer()

for i in alldatatokens:
    new = lemma.lemmatize(i)
    alldatalemma.append(new)
    
alldatatokensdf = pd.DataFrame(np.array(alldatalemma))
alldatatokensdf.columns = ['words']
```

```python
alldatawordlist = alldatatokensdf['words'].value_counts()
alldatareplace = pd.read_csv("adidaswordlist.csv")
alldatareplace.head()
```

```python
replace = alldatareplace['ATTRIBUTE'].tolist()
search = alldatareplace['WORD'].tolist()
replaceDict = dict(zip(search, replace))

def replaceWord(l):
    new_list = []
    for x in l:
        if x not in list(replaceDict.keys()):
            x = x
        else:
            x = replaceDict[x]
        new_list.append(x)
    return new_list
    
alldata['finaltokens'] = alldata['tokens'].apply(replaceWord)
alldata.head()
```

```python
listcomments= alldata['finaltokens'].tolist()
shoes = ['nike', 'adidas', 'newbalance', 'asics']
attributes = ['fit', 'comfortable', 'uncomfortable', 'support', 'athletic', 'style', 'price', 'quality', 'badquality', 'value']
dfattributelift = pd.DataFrame(index = attributes, columns = shoes)


def twobrandcount(brand1, attribute, clist):
    count = 0
    for eachcomment in clist:
        if eachcomment.count(brand1) != 0 and eachcomment.count(attribute) != 0:
            count += 1
    return count

def singlebrandcount(brand, clist):
    count = 0
    for eachcomment in clist:
        if eachcomment.count(brand) != 0:
            count += 1
    return count

def liftratio(brand1, attribute, clist):
    numerator = twobrandcount(brand1, attribute, clist)*len(clist)
    denominator = singlebrandcount(brand1, clist)*singlebrandcount(attribute, clist)
    lift = float(numerator)/denominator
    return lift

for i in shoes:
    for j in attributes:
        dfattributelift.set_value(i, j, liftratio(i,j, listcomments))
```
